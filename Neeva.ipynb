```
from sklearn.model_selection import train_test_split
import pandas as pd


df_sa = pd.read_csv('shared_articles.csv')
df_ui = pd.read_csv('users_interactions.csv')
df = df_ui

df['COUNTER'] =1       #initially, set that counter to 1.
group_data = df.groupby(['contentId','eventType'])['COUNTER'].sum().reset_index() #sum function
print(group_data)

events_df = group_data.pivot_table('COUNTER', ['contentId'], 'eventType')

events_df = events_df.fillna(0)
```

```
def label(row):
   return (1* row['VIEW']) + (4*row['LIKE']) + (10*row['COMMENT CREATED']) +( 25*row['FOLLOW'] )+ (100*row['BOOKMARK'])

events_df['verality'] = events_df.apply (lambda row: label(row), axis=1)
```

```
events_df.describe()
```

```
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
```

```
train, test = train_test_split(events_df, test_size=0.2)

train_X, train_Y = train.drop('verality',axis = 1), train['verality']
test_X, test_Y = test.drop('verality',axis = 1), test['verality']

print(train.head)
```

```
mlp = MLPRegressor(hidden_layer_sizes=(50, 50),
                                 learning_rate_init=0.01,
                                 early_stopping=True)
mlp.fit(train_X,train_Y)
mlp.score(train_X,train_Y)
predict_test = mlp.predict(test_X)

#print(predict_test)
#print(mlp.coefs_)
print(mlp.score)
```
